```python
group_marks_data['math'] = pd.to_numeric(group_marks_data['math'], errors='coerce')
group_marks_data['group'] = group_marks_data['group'].fillna(group_marks_data['group'].mode()[0])
group_marks_data['math'] = group_marks_data['math'].fillna(group_marks_data['math'].mean())
group_marks_data['english'] = group_marks_data['english'].fillna(group_marks_data['english'].mean())
```

```python
group_marks_data.drop_duplicates(inplace=True)
```

```python
group_marks_data = pd.get_dummies(group_marks_data, columns=['gender'], drop_first=True)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
group_marks_data[['age', 'scholarship', 'math', 'english', 'urdu']] = scaler.fit_transform(
    group_marks_data[['age', 'scholarship', 'math', 'english', 'urdu']])
```

```python
cleaned_file_path = 'group_marks_cleaned.csv'
group_marks_data.to_csv(cleaned_file_path, index=False)
```
